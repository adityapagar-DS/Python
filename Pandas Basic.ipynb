```python
#Series
```

```python
#Creating serise using tuple
list = [11,22,3,45,1,6]
s1=pd.Series(list)
s1
```

```python
#Creating serise using tuple
list = [11,22,3,45,1,6]
s1=pd.Series(list,index=['a','b','c','d','e','f'])
s1
```

```python
#Generate serise using tuple and list
list = (11,22,3,45,1,6)
tuple = [22,33,44,55,66,7]
s1=pd.Series(list,index=tuple)
s2=pd.Series(tuple,index=list)
print(s1)
print(s2)
```

```python
#Using Dictonary
ap={'a':1,'b':2,'c':3}
s3=pd.Series(ap)
s3
```

```python
#Using Dictonary
ap={'a':[1,2,3],'b':[2,3,4],'c':[3,4,5]}
s3=pd.Series(ap)
s3
```

```python
ap=[1,2,3,4,5]
yg=[6,7,8,9,10]
s4=pd.Series(ap,yg)
s4
```
